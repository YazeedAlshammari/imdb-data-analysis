#Tidiness Issue 2: Multiple Variables in One Column.

```python
# we have Tidiness issue in df1, where the stars names are in 1 columns insted of being sperated like df2.
#It is better to sperated for easy visual. 
```

```python
#solution 2:
# Split the 'Stars' column into a list of stars
df1['Stars'] = df1['Stars'].str.split(',')

# Limit the number of stars to 4
df1['Stars'] = df1['Stars'].apply(lambda x: x[:4])

# Create new columns for stars
star_columns = [f'Star{i}' for i in range(1, 5)]
df1[star_columns] = pd.DataFrame(df1['Stars'].to_list(), index=df1.index)

# Drop the original 'Stars' column
df1.drop(['Stars'], axis=1, inplace=True)

print(df1)
```

# Answering research Questions:

#Question 1: Analysis of the No_of_votes of a movie vs directors.

```python
director_votes = combined_df.groupby('Director')['No_of_votes'].sum().sort_values(ascending=False)

# Plot the results
plt.figure(figsize=(12, 6))
director_votes.head(10).plot(kind='bar', color='lightgreen')
plt.title('Top 10 Directors by Total Number of Votes')
plt.xlabel('Director')
plt.ylabel('Total Number of Votes')
plt.show()
```

# inshight: we see that Christopher Nolan has the highest voted movies.

#Question 2: What are top 5 IMDB ranking Movies? 

```python
top_5_movies = combined_df.sort_values(by='IMDB_Rating', ascending=False).head(10)

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(top_5_movies['Movie_Name'], top_5_movies['IMDB_Rating'], color='skyblue')
plt.xlabel('Movie Name')
plt.ylabel('IMDb Rating')
plt.title('Top 5 Movies with Highest IMDb Ratings')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# Insight: we found the top 5 imdb rating movies as showed in the figure.
